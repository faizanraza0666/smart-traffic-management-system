```python
# Setting up detectron2 logger
import detectron2
from detectron2.utils.logger import setup_logger
setup_logger()

import numpy as np
import cv2
import random
from google.colab.patches import cv2_imshow
import matplotlib.pyplot as plt

# detectron2 utilities
from detectron2 import model_zoo
from detectron2.engine import DefaultPredictor
from detectron2.config import get_cfg
from detectron2.utils.visualizer import Visualizer
from detectron2.data import MetadataCatalog
import tensorflow as tf
```

```python
cfg = get_cfg()

cfg.merge_from_file(model_zoo.get_config_file("COCO-Detection/faster_rcnn_R_50_C4_3x.yaml"))
cfg.MODEL.ROI_HEADS.SCORE_THRESH_TEST = 0.65  # threshold value for the model

cfg.MODEL.WEIGHTS = model_zoo.get_checkpoint_url("COCO-Detection/faster_rcnn_R_50_C4_3x.yaml")
predictor = DefaultPredictor(cfg)
```

```python
img = cv2.imread("/content/drive/MyDrive/frames/70.png")

outputs = predictor(img)
```

```python
def change_2_red(img,person,p1,p2):
  _ = cv2.rectangle(img, p1, p2, (255,0,0), 2)  
  return img
```

```python
import os
import re

names=os.listdir('/content/drive/MyDrive/frames/')
names.sort(key=lambda f: int(re.sub('\D', '', f)))
```

```python
def prepare(img_array):
  IMG_SIZE = 224

  new_array = cv2.resize(img_array, (IMG_SIZE, IMG_SIZE))
  return new_array.reshape(-1, IMG_SIZE, IMG_SIZE, 3)

def find_closest_people(name,thresh):
  img = cv2.imread('/content/drive/MyDrive/frames/'+name)
  outputs = predictor(img)
  classes=outputs['instances'].pred_classes.cpu().numpy()
  bbox=outputs['instances'].pred_boxes.tensor.cpu().numpy()
  ind = np.where(classes==7)[0]
  person=bbox[ind]
  num= len(person)
  points=[]
  results = []
  model = tf.keras.models.load_model("/content/drive/MyDrive/Colab Notebooks/Vehicle_Detection_final/model_VGG.h5")

  emergency_or_not = ["Emergency", "Non_Emergency"]

  for i in range(num):
    x1,y1,x2,y2 = person[i]
    x1,y1 = int(x1),int(y1)
    x2,y2 = int(x2),int(y2)
    
    test=[]

    test.append([prepare(img[y1:y2, x1:x2])])
    prediction = model.predict(test)
    points.append((x1,y1,x2,y2))
    results.append(emergency_or_not[int(prediction[0][0])])

  for i in range(len(results)):
    x1 = points[i][0]
    y1 = points[i][1]
    x2 = points[i][2]
    y2 = points[i][3]
    if results[i] == 'Emergency':
      img = change_2_red(img,person,(x1,y1),(x2,y2))
  cv2.imwrite('/content/drive/MyDrive/frames/'+name,img)
  return 0
```

```python
from tqdm import tqdm
thresh=100
_ = [find_closest_people(names[i],thresh) for i in tqdm(range(0, 20)) ]
# _ = [find_closest_people(names[i],thresh) for i in tqdm(range(len(names))) ]
```
